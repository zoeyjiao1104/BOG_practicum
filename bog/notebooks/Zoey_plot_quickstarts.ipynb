```python
# Initialize map
map_center = [48.473139, -58.595852] # longitude, latitude
map = folium.Map(location=map_center, zoom_start=13, tiles='openstreetmap', attr="MapQuest", max_zoom=100)

# Compose points
points = list(zip(maine_drifter_2030.latitude.tolist(), maine_drifter_2030.longitude.tolist()))

# Draw line connecting points on map
folium.PolyLine(points, color="blue", weight=2.5, opacity=1).add_to(map)

# Add buoy readings as points on map
for p in points:
  folium.CircleMarker(p, color="red", fill=True, fill_color="red", fill_opacity=1, radius=1).add_to(map)

# Render map (NOTE: You can zoom in and out, move around, export the map to a file, etc)
map
```
